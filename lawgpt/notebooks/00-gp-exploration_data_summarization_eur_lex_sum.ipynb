# first exploration of the data for summarization

```python
import os
```

```python
ROOT_DATA_PATH = "../data/raw/summarization/eur-lex-sum_it/"
```

```python
test_path = os.path.join(ROOT_DATA_PATH, "test.json")
print(os.path.exists(test_path))
print(test_path)
```

```python
import json

data_list = []

with open(test_path, "r") as file:
    for line in file:
        try:
            json_obj = json.loads(line)
            data_list.append(json_obj)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

# Now data_list contains all the JSON objects read from the file
# You can further process or analyze these objects as needed
```
