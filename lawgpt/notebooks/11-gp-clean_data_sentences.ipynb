```python
import re

def clear_text(text):
    # Replace '\r' and '\n' with an empty string
    cleaned_text = text.replace('\\r', '').replace('\\n', '')
    return cleaned_text

def extract_text_between_hashtags(text):
    # Define the pattern
    cleaned_text = clear_text(text)
    pattern = r'### Risposta:\n(.*?)###'
    
    # Search for the pattern in the text
    match = re.search(pattern, cleaned_text, re.DOTALL)
    
    if match:
        return match.group(1)
    else:
        return None

# Example usage:
text = "This is some text ### Risposta This is the text I want to extract ### Prova This is some other text #"
extracted_text = extract_text_between_hashtags(text)
print(extracted_text)
```
