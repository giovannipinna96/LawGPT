```python
import os


def group_files_by_number(root_folder, subfolder):
    file_dict = {}
    file_list = os.path.join(root_folder, subfolder)
    for file_name in os.listdir(file_list):
        if file_name.endswith(".txt"):
            file_number = "".join(
                filter(str.isdigit, file_name)
            )  # Extract numbers from filename
            if file_number not in file_dict:
                file_dict[file_number] = []

            with open(os.path.join(file_list, file_name), "r") as file:
                content = file.read()
                file_dict[file_number].append(content)

    return file_dict


# Example usage:
# file_list = os.path.join(ROOT_FOLDER, "OP")

result_OP = group_files_by_number(ROOT_FOLDER, "OP")
print(result_OP)
```

```python
import os


def read_files_in_folders(root_folder):
    data_dict = {}

    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            for file_name in os.listdir(folder_path):
                if file_name.endswith(".txt"):
                    file_path = os.path.join(folder_path, file_name)
                    # Extract the key from file name
                    key = file_name.split("-")[0]

                    # ! questo è il tipo di encoding corretto
                    with open(file_path, "r", encoding="utf-8-sig") as file:
                        content = file.read().strip()

                        if key not in data_dict:
                            data_dict[key] = [content]
                        else:
                            data_dict[key].append(content)

    return data_dict


for sub_folder in sub_path:
    result = read_files_in_folders(ROOT_FOLDER)
    # print(result)
```

```python
def read_file(file_path):
    try:
        with open(file_path, "r") as file:
            lines = file.readlines()
            return lines
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


# Example usage:
# file_path = "path/to/your/file.txt"  # Replace with your file path
# file_content = read_file(file_path)

# if file_content is not None:
#     print(file_content)  # Display the content of the file if it's not None
```

```python
def union_lists(*lists):
    # result = []
    # for i in range(len(list1)):
    #     result.append([list1[i], list2[i]])
    # return result

    result = []
    for i in range(len(lists[0])):
        sublist = []
        for lst in lists:
            sublist.append(lst[i])
        result.append(sublist)
    return result
```

```python
import xml.etree.ElementTree as ET


def parse_xml_file(file_path):
    result = []

    tree = ET.parse(file_path)
    root = tree.getroot()

    for simplification in root.findall(".//simplification"):
        simplification_data = {}
        simplification_data["type"] = simplification.get("type")
        simplification_data["origin"] = simplification.get("origin")

        before_text = simplification.find("before").text
        simplification_data["before"] = (
            before_text.strip() if before_text is not None else ""
        )

        after_text = simplification.find("after").text
        simplification_data["after"] = (
            after_text.strip() if after_text is not None else ""
        )

        result.append(simplification_data)

    return result


# Example usage:
file_path = "simpitiki-v2.xml"
parsed_data = parse_xml_file(os.path.join(ROOT_FOLDER, file_path))
for entry in parsed_data:
    print(entry)
```

```python
conversion_type_dict = {
    "1": "Split",
    "2": "Merge",
    "3": "Reordering",
    "11": "Insert - Verb",
    "12": "Insert - Subject",
    "13": "Insert - Other",
    "21": "Delete - Verb",
    "22": "Delete - Subject",
    "23": "Delete - Other",
    "31": "Transformation - Lexical Substitution (word level)",
    "32": "Transformation - Lexical Substitution (phrase level)",
    "33": "Transformation - Anaphoric replacement",
    "34": "Transformation - Noun to Verb",
    "35": "Transformation - Verb to Noun (nominalization)",
    "36": "Transformation - Verbal Voice",
    "37": "Transformation - Verbal Features",
}
```

```python
def replace_dict_value(dict1, dict2, key):
    if dict1[key] in dict2.keys():
        dict1[key] = dict2[dict1[key]]
    return dict1
```

```python
def replace_lst_of_dict_value(list_of_dict, dict2, key):
    return [replace_dict_value(ld, dict2, key) for ld in list_of_dict]
```

```python
final_simpliwoki_dict = replace_lst_of_dict_value(
    parsed_data, conversion_type_dict, "type"
)
```
